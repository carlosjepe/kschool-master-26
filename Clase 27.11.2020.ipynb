```python
def process_b(lines):
    
    result = []
    
    for line in lines:
        if len(line) > 1:
            result.append(line[0])
    return result

process_b(lines)
```

```python
def process(f, lines):

    result = []

    for line in lines:
        if len(line) > 1:
            result.append(f(line))

    return result

process(len, lines)
```

# List Comprehensions

### Son basicamente una forma compacta de escribir un "for" o un "map". Se usan corchetes porque son listas

```python
[ n ** 2 for n in my_list ]
```

```python
result = []

for n in my_list:
    result.append(n ** 2)
    
result
```

```python
list(map(lambda n: n ** 2, my_list))
```

### Son tres formas de hacer lo mismo (lo que está arriba)
